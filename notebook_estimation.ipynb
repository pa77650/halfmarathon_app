````python
df_model = df[
    ['Płeć', 'Rocznik', '5 km Czas', 'Czas']
].copy()

df_model = df_model.dropna()

print(df_model.shape)
````

````python
df_model.isna().sum()
````

````python
df_model[['5 km Czas', 'Czas']]
````

````python
df_model["time_5km_sec"] = (
    pd.to_timedelta(df_model["5 km Czas"])
    .dt.total_seconds()
)

df_model["time_half_sec"] = (
    pd.to_timedelta(df_model["Czas"])
    .dt.total_seconds()
)

df_model[
    ["5 km Czas", "time_5km_sec",
     "Czas", "time_half_sec"]
].head()

````

````python
df_model = df_model.rename(columns={
    "5 km Czas": "time_5km",
    "Czas": "time_half",
    "Rocznik": "year_birth",
    "Płeć": "gender"
})
````

````python
CURRENT_YEAR = 2024

df_model["age"] = CURRENT_YEAR - df_model["year_birth"]
````

````python
df_model = df_model[
    (df_model["age"] >= 18) &
    (df_model["age"] <= 90) &
    (df_model["time_half_sec"] > 0)
].copy()

print(df_model.shape)
````

````python
df_model[
    ["age", "time_5km_sec", "time_half_sec"]
].describe()
````

````python
df_model["gender_"] = df_model["gender"].map({
    "M": 0,
    "K": 1
})
````

````python
df_model[["gender", "gender_"]]
````

````python
model_data = df_model[
    [
        "gender",
        "age",
        "time_5km_sec",
        "time_half_sec"
    ]
].copy()
````

````python
model_data
````

````python
from pycaret.regression import *
````

````python
reg_setup = setup(
    data=model_data,
    target="time_half_sec",
    session_id=42,
    verbose=False
)
````

````python
best_model = compare_models()
````

````python
final_model = finalize_model(best_model)
````

````python
save_model(
    final_model,
    "models/halfmarathon_model"
)
````

````python
from openai import OpenAI
from dotenv import load_dotenv
import os
import json

load_dotenv(override=True)

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

user_text = """
Mam 30 lat, jestem kobietą, 5 km biegnę w 25 minut.

"""

response = client.responses.create(

    model="gpt-4o-mini",
    input=f"""

Wyłuskaj dane z tekstu i zwróć tylko i wyłącznie JSON:

{{
  "gender": 0,
  "age": 0,
  "time_5km_sec": 0
}}

Oznaczenie płci: Kobieta 1, Mężczyzna 0
Czas na 5 km zawsze zwróć w sekundach, niezależnie od formatu wejściowego.
Jeśli brakuje danych zwróć null dla danego pola.
Jeśli wiek (age) powyżej 120 lat, to upewnij się o poprawność danych.

Tekst:
{user_text}
"""
)

print(response.output_text)
````

````python
import json
import pandas as pd
from pycaret.regression import predict_model

clean_json = (
    response.output_text
    .replace("```json", "")
    .replace("```", "")
    .strip()
)

data = json.loads(clean_json)

test_runner = pd.DataFrame({
    "gender": [data["gender"]],
    "age": [data["age"]],
    "time_5km_sec": [data["time_5km_sec"]]
})

prediction = predict_model(
    final_model,
    data=test_runner
)

prediction[["prediction_label"]]
````
